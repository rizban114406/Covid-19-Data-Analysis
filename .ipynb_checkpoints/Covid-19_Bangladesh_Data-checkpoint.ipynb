```python
# importing libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
#Load Csv file
caseBangladesh =pd.read_csv('C://Users/Faisal-Aakon/Desktop/caseBangladesh.csv')
coronaSituationBd =pd.read_csv('C://Users/Faisal-Aakon/Desktop/coronaSituationBd.csv')
ageRangeCaseBangladesh =pd.read_csv('C://Users/Faisal-Aakon/Desktop/ageRangeCaseBangladesh.csv')
```

```python
coronaSituationBd
```

```python
#Pie ploting male vs female affected patient 
df = pd.DataFrame({coronaSituationBd.Male.sum(),coronaSituationBd.Female.sum()
                   },
                  index=['Female', 'Male'])
my_explode=(0,0.2)
df.columns=['Male_VS_Female']
df.plot.pie(subplots=True,figsize=(10, 10),autopct='%1.1f%%',shadow= True,explode=my_explode, startangle=90)
```

```python
#Pie ploting UnderTreatment vs Cured vs Dead patient 
df = pd.DataFrame({coronaSituationBd.UnderTreatment.sum(),coronaSituationBd.Cured.sum(),
                   coronaSituationBd.Dead.sum()
                   },
                  index=['UnderTreatment', 'Cured','Dead'])
my_explode=(0,0,0.2)
df.columns=['UnderTreatment_VS_Cured_VS_Dead']
df.plot.pie(subplots=True,figsize=(10, 10),autopct='%2.1f%%',shadow= True,explode=my_explode, startangle=90)
```

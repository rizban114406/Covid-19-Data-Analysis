```python
# importing libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
#Load Csv file
caseBangladesh =pd.read_csv('../Analysis Data/caseBangladesh.csv')
coronaSituationBd =pd.read_csv('../Analysis Data/coronaSituationBd.csv')
ageRangeCaseBangladesh =pd.read_csv('../Analysis Data/ageRangeCaseBangladesh.csv')
districtWiseCase =pd.read_csv('../Analysis Data/DistrictWiseCase.csv')
testCaseNo =pd.read_csv('../Analysis Data/testCaseNo.csv')
```

```python
coronaSituationBd
```

```python
#Pie ploting male vs female affected patient 
df = pd.Series([coronaSituationBd.Male.sum(),coronaSituationBd.Female.sum()])
df.index=['Male','Female']
my_explode=(0,0.2)
df.columns=['Male_VS_Female']
df.plot.pie(subplots=True,figsize=(10, 10),autopct='%1.1f%%',shadow= True,explode=my_explode, startangle=90)
```

```python
#Pie ploting UnderTreatment vs Cured vs Dead patient 
# df = pd.DataFrame({coronaSituationBd.UnderTreatment.sum(),coronaSituationBd.Cured.sum(),
#                    coronaSituationBd.Dead.sum()
#                    },
#                   index=['UnderTreatment', 'Cured','Dead'])
df=pd.Series([coronaSituationBd.UnderTreatment.sum(),coronaSituationBd.Cured.sum(),
                   coronaSituationBd.Dead.sum()])
df.index=['UnderTreatment', 'Cured','Dead']

my_explode=(0,0,0.2)
df.columns=['UnderTreatment_VS_Cured_VS_Dead']
df.plot.pie(subplots=True,figsize=(10, 10),autopct='%2.1f%%',shadow= True,explode=my_explode, startangle=90)
```

```python
#Day wise number of case
# plt.bar(caseBangladesh.day,caseBangladesh.case,color='#800000',figsize=(20,5))

# plt.title('day wise case')
# plt.ylabel('case')
# plt.xlabel('day')
# plt.show()
fig = plt.figure(figsize=(20,5))
ax1 = plt.subplot2grid((1,1), (0,0))
x=caseBangladesh.day
y=caseBangladesh.case
ax1.bar(x,y, label='Bar1', color='#800000')

for label in ax1.xaxis.get_ticklabels(): # loop to change the x axis label rotation
    label.set_rotation(90)
# for index, value in enumerate(y):
#     plt.text(value, index, str(value))

plt.title('day wise case')
plt.ylabel('case')
plt.xlabel('day')
plt.legend()

plt.show()
```

```python
fig = plt.figure(figsize=(20,5))
ax1 = plt.subplot2grid((1,1), (0,0))
y=df['case']
x=df['City']
ax1.bar(x,y, label='Bar1', color='#800000')

for label in ax1.xaxis.get_ticklabels(): # loop to change the x axis label rotation
    label.set_rotation(90)
# for index, value in enumerate(y):
#     plt.text(value, index, str(value))

plt.xlabel('District')
plt.ylabel('Number Of Case')
plt.title("District wise Cases")
plt.legend()

plt.show()
```
